```python
import pandas as pd
```

```python
df = pd.read_csv("^DJI.csv")
```

```python
df = df.rename(columns = {'Date':'datadate', 'Open':'prcod', 'High':'prchd', 'Low':'prcld', 'Close':'prccd', 'Volume':'cshtrd'})
```

```python
df['datadate'] = pd.to_datetime(df['datadate'])
df['datadate'] = df.datadate.dt.strftime('%Y%m%d')
```

```python
df['ajexdi'] = df['Adj Close']/df['prccd']
```

```python
df['tic'] = 'DJI'
```

```python
df = df.drop(columns = ['Adj Close'])
df = df[['datadate', 'tic', 'prccd', 'prcod', 'prchd', 'prcld', 'ajexdi', 'cshtrd']]
```
