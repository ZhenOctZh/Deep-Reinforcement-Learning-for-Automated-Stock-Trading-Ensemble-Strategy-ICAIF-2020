```python
import pandas as pd
```

```python
df = pd.read_csv("done_data.csv")
```

```python
df
```

```python
df = df[df.tic=='DJI']
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df
```

```python
for i in range(1, 21):
    for col in ['adjcp', 'open', 'high', 'low', 'volume', 'macd', 'rsi', 'cci', 'adx', 'turbulence']:
        df[col+str(i)] = df[col].shift(i, fill_value='-' )
```

```python
df[20:].to_csv('done_data_rolling_window.csv')
```
